```python
# Inventory Optimization Analysis
print("ðŸ“¦ INVENTORY OPTIMIZATION & EFFICIENCY ANALYSIS")
print("=" * 60)

# Inventory health by category
inventory_analysis = client.query("""
SELECT 
  p.category,
  COUNT(i.product_id) as products_count,
  ROUND(AVG(i.available_qty), 0) as avg_stock_level,
  ROUND(AVG(i.total_demand), 0) as avg_demand,
  ROUND(AVG(i.available_qty / NULLIF(i.total_demand, 0)), 2) as avg_stock_ratio,
  ROUND(SUM(i.inventory_value), 2) as total_inventory_investment,
  COUNT(CASE WHEN i.stock_status = 'Excess' THEN 1 END) as excess_stock_products,
  COUNT(CASE WHEN i.stock_status = 'Low' THEN 1 END) as low_stock_products,
  COUNT(CASE WHEN i.stock_status = 'Optimal' THEN 1 END) as optimal_stock_products
FROM `demand-analytics-4482.Planning_data.Inventory` i
JOIN `demand-analytics-4482.Planning_data.Products` p ON i.product_id = p.product_id
GROUP BY p.category
ORDER BY total_inventory_investment DESC
""").to_dataframe()

print("INVENTORY EFFICIENCY BY CATEGORY:")
display(inventory_analysis)

# Inventory optimization opportunities
inventory_optimization = client.query("""
SELECT 
  i.product_id,
  p.product_name,
  p.category,
  i.available_qty,
  i.total_demand,
  i.stock_status,
  ROUND(i.inventory_value, 2) as inventory_value,
  ROUND(i.available_qty / NULLIF(i.total_demand, 0), 2) as stock_coverage_ratio,
  CASE 
    WHEN i.available_qty > i.total_demand * 2 THEN 'Reduce Inventory'
    WHEN i.available_qty < i.total_demand * 0.8 THEN 'Increase Inventory'
    ELSE 'Maintain Current Level'
  END as recommendation
FROM `demand-analytics-4482.Planning_data.Inventory` i
JOIN `demand-analytics-4482.Planning_data.Products` p ON i.product_id = p.product_id
ORDER BY inventory_value DESC
""").to_dataframe()

print("\nINVENTORY OPTIMIZATION RECOMMENDATIONS:")
display(inventory_optimization.head(15))

# Calculate total carrying cost savings potential
excess_inventory_value = inventory_optimization[
    inventory_optimization['recommendation'] == 'Reduce Inventory'
]['inventory_value'].sum()

print(f"\nðŸ’° FINANCIAL IMPACT:")
print(f"   â€¢ Total Inventory Investment: R{inventory_analysis['total_inventory_investment'].sum():,.2f}")
print(f"   â€¢ Excess Inventory Value: R{excess_inventory_value:,.2f}")
print(f"   â€¢ Potential Carrying Cost Savings: R{excess_inventory_value * 0.25:,.2f} (25% carrying cost)")
```

```python
# Forecast Accuracy & Predictive Analysis
print("ðŸ”® FORECAST ACCURACY & PREDICTIVE INSIGHTS")
print("=" * 60)

# Forecast accuracy by type and confidence level
forecast_performance = client.query("""
SELECT 
  forecast_type,
  confidence_level,
  COUNT(*) as total_forecasts,
  ROUND(AVG(variance_pct), 2) as avg_variance_pct,
  ROUND(STDDEV(variance_pct), 2) as variance_std_dev,
  ROUND(AVG(ABS(variance_pct)), 2) as mean_absolute_error,
  ROUND(SQRT(AVG(POW(variance_pct, 2))), 2) as root_mean_square_error,
  COUNT(CASE WHEN ABS(variance_pct) <= 10 THEN 1 END) as accurate_forecasts,
  ROUND(COUNT(CASE WHEN ABS(variance_pct) <= 10 THEN 1 END) * 100.0 / COUNT(*), 2) as accuracy_rate_10pct,
  COUNT(CASE WHEN ABS(variance_pct) <= 20 THEN 1 END) as acceptable_forecasts,
  ROUND(COUNT(CASE WHEN ABS(variance_pct) <= 20 THEN 1 END) * 100.0 / COUNT(*), 2) as accuracy_rate_20pct
FROM `demand-analytics-4482.Planning_data.Forecasts`
GROUP BY forecast_type, confidence_level
ORDER BY accuracy_rate_10pct DESC
""").to_dataframe()

print("FORECAST ACCURACY PERFORMANCE:")
display(forecast_performance)

# Forecast bias analysis
forecast_bias = client.query("""
SELECT 
  forecast_type,
  COUNT(CASE WHEN variance_pct > 0 THEN 1 END) as underforecast_count,
  COUNT(CASE WHEN variance_pct < 0 THEN 1 END) as overforecast_count,
  ROUND(COUNT(CASE WHEN variance_pct > 0 THEN 1 END) * 100.0 / COUNT(*), 2) as underforecast_rate,
  ROUND(COUNT(CASE WHEN variance_pct < 0 THEN 1 END) * 100.0 / COUNT(*), 2) as overforecast_rate,
  ROUND(AVG(variance_pct), 2) as avg_bias
FROM `demand-analytics-4482.Planning_data.Forecasts`
GROUP BY forecast_type
""").to_dataframe()

print("\nFORECAST BIAS ANALYSIS:")
display(forecast_bias)

# Best performing forecast methods
best_forecast_method = forecast_performance.loc[
    forecast_performance['accuracy_rate_10pct'].idxmax()
]

print(f"\nðŸŽ¯ FORECAST OPTIMIZATION INSIGHTS:")
print(f"   â€¢ Best Method: {best_forecast_method['forecast_type']} with {best_forecast_method['confidence_level']} confidence")
print(f"   â€¢ Best Accuracy Rate: {best_forecast_method['accuracy_rate_10pct']}% (within 10% variance)")
print(f"   â€¢ Average Absolute Error: {best_forecast_method['mean_absolute_error']}%")
print(f"   â€¢ Improvement Opportunity: Switch underperforming forecasts to best method")
```

```python
# Sales Event ROI & Investment Analysis
print("ðŸ’¡ SALES EVENT ROI & INVESTMENT STRATEGY")
print("=" * 60)

# Sales event ROI calculation
sales_event_roi = client.query("""
WITH event_performance AS (
  SELECT 
    sale_event,
    COUNT(*) as total_orders,
    SUM(qty) as total_units,
    ROUND(AVG(discount_pct), 2) as avg_discount_pct,
    ROUND(SUM(order_value), 2) as gross_revenue,
    ROUND(SUM(base_price * qty), 2) as full_price_revenue,
    ROUND(SUM(base_price * qty) - SUM(order_value), 2) as discount_amount,
    DATE_DIFF(MAX(order_date), MIN(order_date), DAY) + 1 as event_duration_days
  FROM `demand-analytics-4482.Planning_data.Orders`
  WHERE is_on_sale = true
  GROUP BY sale_event
)
SELECT 
  sale_event,
  total_orders,
  gross_revenue,
  discount_amount,
  event_duration_days,
  ROUND(gross_revenue / total_orders, 2) as revenue_per_order,
  ROUND(gross_revenue / event_duration_days, 2) as daily_revenue,
  ROUND(discount_amount / gross_revenue * 100, 2) as effective_discount_rate,
  ROUND((gross_revenue - discount_amount * 0.3) / (discount_amount), 2) as estimated_roi
FROM event_performance
ORDER BY gross_revenue DESC
""").to_dataframe()

print("SALES EVENT ROI ANALYSIS:")
display(sales_event_roi)

# Event efficiency metrics
total_promotional_revenue = sales_event_roi['gross_revenue'].sum()
total_discount_investment = sales_event_roi['discount_amount'].sum()

print(f"\nðŸ“Š EVENT PORTFOLIO PERFORMANCE:")
print(f"   â€¢ Total Promotional Revenue: R{total_promotional_revenue:,.2f}")
print(f"   â€¢ Total Discount Investment: R{total_discount_investment:,.2f}")
print(f"   â€¢ Portfolio ROI: {(total_promotional_revenue / total_discount_investment):.2f}x")
print(f"   â€¢ Average Event Duration: {sales_event_roi['event_duration_days'].mean():.1f} days")

# Top performing events for investment
top_3_events = sales_event_roi.nlargest(3, 'estimated_roi')
print(f"\nðŸ† TOP 3 EVENTS FOR INCREASED INVESTMENT:")
for idx, event in top_3_events.iterrows():
    print(f"   {idx+1}. {event['sale_event']}: ROI {event['estimated_roi']:.1f}x, Revenue R{event['gross_revenue']:,.0f}")
```

```python
# Cell 15: Create Optimized Views for Looker Studio Dashboard (Correct Names)
from google.cloud import bigquery
import pandas as pd
from datetime import datetime

print("ðŸ”„ Creating optimized data views for Looker Studio dashboard...")
print("ðŸŒ Region: Belgium")

# Initialize BigQuery client
client = bigquery.Client(project='demand-analytics-4482')

print("ðŸ“Š Creating executive KPI summary view...")

# Create Executive KPI Summary View
executive_kpi_query = """
CREATE OR REPLACE VIEW `demand-analytics-4482.Planning_data.executive_kpi_view` AS
SELECT 
    'Current Period' as period_type,
    COUNT(*) as total_orders,
    SUM(order_value) as total_revenue,
    AVG(order_value) as avg_order_value,
    COUNT(DISTINCT product_id) as active_products,
    COUNT(DISTINCT region) as active_regions,
    COUNT(DISTINCT customer_type) as customer_segments,
    SUM(CASE WHEN is_on_sale = TRUE THEN order_value ELSE 0 END) as promotional_revenue,
    SUM(CASE WHEN is_on_sale = TRUE THEN order_value ELSE 0 END) / SUM(order_value) * 100 as promotional_share_pct
FROM `demand-analytics-4482.Planning_data.Orders`
"""

# Create Category Performance View
category_performance_query = """
CREATE OR REPLACE VIEW `demand-analytics-4482.Planning_data.category_performance_view` AS
SELECT 
    category,
    COUNT(*) as orders,
    SUM(order_value) as revenue,
    AVG(order_value) as avg_order_value,
    SUM(order_value) / (SELECT SUM(order_value) FROM `demand-analytics-4482.Planning_data.Orders`) * 100 as revenue_share_pct,
    SUM(CASE WHEN is_on_sale = TRUE THEN 1 ELSE 0 END) as promotional_orders,
    SUM(CASE WHEN is_on_sale = TRUE THEN 1 ELSE 0 END) / COUNT(*) * 100 as promotional_rate_pct
FROM `demand-analytics-4482.Planning_data.Orders`
GROUP BY category
ORDER BY revenue DESC
"""

# Create Regional Performance View
regional_performance_query = """
CREATE OR REPLACE VIEW `demand-analytics-4482.Planning_data.regional_performance_view` AS
SELECT 
    region,
    COUNT(*) as orders,
    SUM(order_value) as revenue,
    AVG(order_value) as avg_order_value,
    SUM(order_value) / (SELECT SUM(order_value) FROM `demand-analytics-4482.Planning_data.Orders`) * 100 as revenue_share_pct,
    COUNT(DISTINCT customer_type) as customer_segments,
    COUNT(DISTINCT product_id) as unique_products
FROM `demand-analytics-4482.Planning_data.Orders`
GROUP BY region
ORDER BY revenue DESC
"""

# Create Time Series View for Trends
time_series_query = """
CREATE OR REPLACE VIEW `demand-analytics-4482.Planning_data.time_series_view` AS
SELECT 
    order_month,
    COUNT(*) as orders,
    SUM(order_value) as revenue,
    AVG(order_value) as avg_order_value,
    SUM(CASE WHEN is_on_sale = TRUE THEN order_value ELSE 0 END) as promotional_revenue,
    COUNT(DISTINCT product_id) as active_products,
    COUNT(DISTINCT customer_type) as customer_segments
FROM `demand-analytics-4482.Planning_data.Orders`
GROUP BY order_month
ORDER BY order_month
"""

# Create Customer Segmentation View
customer_segmentation_query = """
CREATE OR REPLACE VIEW `demand-analytics-4482.Planning_data.customer_segmentation_view` AS
SELECT 
    customer_type,
    COUNT(*) as orders,
    SUM(order_value) as revenue,
    AVG(order_value) as avg_order_value,
    SUM(order_value) / (SELECT SUM(order_value) FROM `demand-analytics-4482.Planning_data.Orders`) * 100 as revenue_share_pct,
    COUNT(DISTINCT region) as regions_served,
    COUNT(DISTINCT category) as categories_purchased
FROM `demand-analytics-4482.Planning_data.Orders`
GROUP BY customer_type
ORDER BY revenue DESC
"""

# Create Inventory Optimization View
inventory_optimization_query = """
CREATE OR REPLACE VIEW `demand-analytics-4482.Planning_data.inventory_optimization_view` AS
SELECT 
    stock_status,
    COUNT(*) as product_count,
    SUM(inventory_value) as total_inventory_value,
    AVG(inventory_value) as avg_inventory_value,
    SUM(available_qty) as total_quantity,
    SUM(total_demand) as total_demand,
    SUM(total_demand) / SUM(available_qty) as turnover_ratio,
    SUM(inventory_value) / (SELECT SUM(inventory_value) FROM `demand-analytics-4482.Planning_data.Inventory`) * 100 as value_share_pct
FROM `demand-analytics-4482.Planning_data.Inventory`
GROUP BY stock_status
ORDER BY total_inventory_value DESC
"""

# Create Forecast Performance View
forecast_performance_query = """
CREATE OR REPLACE VIEW `demand-analytics-4482.Planning_data.forecast_performance_view` AS
SELECT 
    forecast_type,
    confidence_level,
    COUNT(*) as forecast_count,
    AVG(ABS(variance_pct)) as avg_abs_variance,
    STDDEV(variance_pct) as variance_std_dev,
    100 - AVG(ABS(variance_pct)) as accuracy_pct,
    SUM(CASE WHEN ABS(variance_pct) <= 10 THEN 1 ELSE 0 END) / COUNT(*) * 100 as within_10pct_accuracy
FROM `demand-analytics-4482.Planning_data.Forecasts`
GROUP BY forecast_type, confidence_level
ORDER BY accuracy_pct DESC
"""

# Execute all view creation queries
try:
    print("ðŸ”„ Creating BigQuery views with correct names...")
    
    client.query(executive_kpi_query).result()
    print("âœ… Executive KPI view created")
    
    client.query(category_performance_query).result()
    print("âœ… Category performance view created")
    
    client.query(regional_performance_query).result()
    print("âœ… Regional performance view created")
    
    client.query(time_series_query).result()
    print("âœ… Time series view created")
    
    client.query(customer_segmentation_query).result()
    print("âœ… Customer segmentation view created")
    
    client.query(inventory_optimization_query).result()
    print("âœ… Inventory optimization view created")
    
    client.query(forecast_performance_query).result()
    print("âœ… Forecast performance view created")
    
    print("\nðŸŽ¯ All BigQuery views created successfully!")
    print("ðŸ“Š Views are now ready for Looker Studio connection")
    
    # Display view summary
    print("\nðŸ“‹ Created Views Summary:")
    print("   â€¢ executive_kpi_view - Key performance indicators")
    print("   â€¢ category_performance_view - Category analysis")
    print("   â€¢ regional_performance_view - Regional breakdown")
    print("   â€¢ time_series_view - Trend analysis")
    print("   â€¢ customer_segmentation_view - Customer insights")
    print("   â€¢ inventory_optimization_view - Inventory metrics")
    print("   â€¢ forecast_performance_view - Forecasting accuracy")
    
except Exception as e:
    print(f"âŒ Error creating views: {e}")

print("\nðŸš€ Ready for Looker Studio connection!")
print("ðŸ“Œ Next step: Connect to Looker Studio and create dashboard")
```
